```
 from tensorflow import keras
import numpy as np
from matplotlib import pyplot as plt
```

```
# load dataset
(x_train, y_train), (x_test, y_test) = keras.datasets.mnist.load_data()
```

```
print('x_train is {}'.format(x_train.shape))
print('y_train is {}'.format(y_train.shape))
print('x_test  is {}'.format(x_test.shape))
print('y_test  is {}'.format(y_test.shape))
```

```
# show some samples of the dataset
plt.figure(figsize=(25,4))
for i in range(8):
    plt.subplot(1,8,i+1)
    plt.imshow(x_train[i], cmap='gray')
    plt.xticks([])
    plt.yticks([])
    plt.title(y_train[i])
```

```
h = np.zeros(10)
for n in range(y_train.shape[0]):
  h[y_train[n]] += 1

print(h)
plt.stem(h)
```

```
rows = 10
cols = 20
h = np.zeros(rows)
image = np.ones((30*rows, 30*cols))*255
for n in range(y_train.shape[0]):
    if h[y_train[n]] < cols:
        r = int(y_train[n])
        c = int(h[y_train[n]])
        image[r*30+1:(r+1)*30-1, c*30+1:(c+1)*30-1] = x_train[n]

        h[y_train[n]] += 1

plt.figure(figsize=(10, 10))
plt.imshow(image, 'gray')
plt.xticks([])
plt.yticks([])
plt.axis('off');
```

```
nums = 10
image = np.ones((30*1, 30*nums))*255
for n in range(nums):
    ind = np.where(y_train == n)
    image[1:29, n*30:(n+1)*30-2] = np.average(x_train[ind], axis=0)

plt.figure(figsize=(10, 10))
plt.imshow(image, 'gray')
plt.xticks([])
plt.yticks([])
plt.axis('off');
```
